### Problem Statement 1: Samantha has created a dataset named 'top50spotify.csv' of her top 50
songs from spotify.
Dataset Description:
top50spotify.csv - The dataset contains 14 features. Here's a brief description of a few columns
in the dataset:
• SerialNo. - Serial number of songs
• Track.Name - Name of the track
• Artist.Name - Name of the artist
• Genre - Genre of the song
• Energy - Energy index of the song
• Length. - Length of the song
• Popularity - Popularity index of the song
Tasks to be performed:
1. Import the dataset as a DataFrame and drop the first column.
2. Save it as 'top50.csv'.
3. Find the average Energy and Length of first 10 songs.
4. Find the total length of songs, group by genre from top to bottom.
5. Print the artist name with the most number of tracks in one genre. (Hint: Group by artist name and genre)
6. Print the data of the tracks created by the artist from the previous question.

```python
# 1. Import the dataset as a DataFrame and drop the first column.
import pandas as pd
data = pd.read_csv("top50spotify.csv" )
data.drop('SerialNo.',axis=1,inplace=True)
data.head()
```

```python
# 2. Save it as 'top50.csv'.
data.to_csv('top50.csv')
```

```python
# 3. Find the average Energy and Length of first 10 songs.
avg1 =data.head(10).Energy.mean()
print("Average Energy of first 10 songs: ",avg1)

avg2 =data.head(10)["Length."].mean()
print("Average Length of first 10 songs: ",avg2)
```

```python
# 4. Find the total length of songs, group by genre from top to bottom.
data.groupby("Genre")["Length."].sum().sort_values(ascending=False)
```

```python
# 5. Print the artist name with the most number of tracks in one genre. (Hint: Group by artist name and genre)
art =data.groupby(["Artist.Name","Genre"]).count()["Track.Name"].sort_values(ascending=False)
art=art.head(1)
art

```

```python
# 6. Print the data of the tracks created by the artist from the previous question.
data[data['Artist.Name']=='Ed Sheeran']
```

### Problem Statement 2: Write a Python program to perform the following 
tasks
1.Create a pandas series from the below dictionary where indices
are subjects:
{'English':{'Sam':60,'Jackson':74,'Ahree':85},
'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61
},
'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'A
hree':76},'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,
'Gray':95},
'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87
}
}

2.Convert the created series into DataFrame and replace the null values with zeroes.

3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by
calculating the average of all subjects.

```python
series = pd.Series({'English':{'Sam':60,'Jackson':74,'Ahree':85},
                    'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
                    'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
                    'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
                    'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87
                    }})

series=series.apply(pd.Series)
series
```

```python
# 2.Convert the created series into DataFrame and replace the null values with zeroes.
series=series.fillna(0)
series
```

```python
# 3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by
# calculating the average of all subjects.

series =series.transpose()
# series["Average"]
series
```

```python
avg=[]
for i in range(series.shape[0]):
    avg.append(round(sum(list(series.iloc[i,:]))/len(series.columns),2))
series["Average"]= avg
series
```

 ### Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select
only numbers divisible by 7 and 17.


```python
lim=list(range(1,1001))
def con(lim):
    if (lim%7==0) & (lim%17 == 0):
        return lim
    
    
ser= pd.Series(list(filter(con,lim)))
ser
```

### Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.

Dataset Description:
cereal.csv - The dataset contains 16 features. Here's a brief description of 3 columns in the
dataset:
• name - Brand name of the cereals
• MFR - Manufacturer of the brands
• rating - Rating of the cereals
Syliphia wants to visualize the quality of cereals and determine which manufacturer delivers
the best quality.
Tasks to be performed:

1. Import the dataset.
2. Plot ratings of different types of manufacturers.
3. Use xticks range form 0-100.
4. Change the style of the graph to seaborn

```python
# 1. Import the dataset.
import pandas as pd
syl_data=pd.read_csv('cereal.csv')
syl_data.head()
```

```python
rat=syl_data.groupby("mfr")["rating"].mean()
rat.index
```

```python
# 2. Plot ratings of different types of manufacturers.
from matplotlib import pyplot as plt
# plt.figure(figsize=(25,5))
plt.style.use("ggplot")
plt.barh(rat.index,rat.values)

```

```python
# 3. Use xticks range form 0-10and0.
plt.barh(rat.index,rat.values)
plt.xlim(0,100);
```

```python
# 4. Change the style of the graph to seaborn
plt.style.use("seaborn")
plt.barh(rat.index,rat.values)
plt.xlim(0,100);
```
